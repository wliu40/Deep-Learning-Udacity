```python
import numpy as np

import pandas as pd

my_data = np.array([[5, 'a', 1],
              [3, 'b', 3],
              [1, 'b', 2],
              [3, 'a', 1],
              [4, 'b', 2],
              [7, 'c', 1],
              [7, 'c', 1]]) 
                   
df = pd.DataFrame(data=my_data, columns=['y', 'dummy', 'x'])

df
```

```python
just_dummies = pd.get_dummies(df['dummy'])
just_dummies
```

```python
step_1 = pd.concat([df, just_dummies], axis=1)

step_1
```

```python
step_1.drop(['dummy', 'c'], inplace=True, axis=1) 
step_1
```
